```python
# Plot decision tree
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data
dataset = pd.read_csv('pima-indians-diabetes.csv', delimiter = ',')
```

```python
# Split data into x and y
y=dataset['1']     # Dependent Variable
X=dataset.drop(['1'],axis=1) # Independent Variables
```

```python
# use train_test_split() function from the scikit-learn library.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =.30, random_state = 123)
```

```python
# train XGBoost Model
from xgboost import XGBClassifier
model = XGBClassifier(use_label_encoder=False)
```

```python
model.fit(X_train, y_train)
```

```python
# plot single tree
from xgboost import plot_tree
#plot_tree(model)
#plt.show()

fig, ax = plt.subplots(figsize=(150, 100))
plot_tree(model, num_trees=2, ax=ax)
fig.savefig('tree.png')
plt.show()
```
